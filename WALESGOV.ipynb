```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
from urllib import request
from urllib.request import Request, urlopen
import pandas as pd
```

```python
def html_loader(url):
    request_site = Request(url, headers={"User-Agent": "Mozilla/5.0"})
    webpage = urlopen(request_site).read()
    soup = BeautifulSoup(webpage,"html.parser")
    return(soup)
```

```python
name = []
party = []
address = []
phone = []
charge = []
email = []
region = []
urls = []
name_party_region_charge = []
sophie = []
```

```python
# Url from Wales Parliament with a list of the MSPs
url = "https://senedd.wales/find-a-member-of-the-senedd/?PageSize=1000&Page=1&Culture=en-GB&ViewModelType=MemberOfSenedd&IsSubSearch=False&CurrentPageId=1441&IsPostcodeCrossConstituency=False&Postcode=&Name=&ShowAll=true&Region=&Constituency=&Constituency=&Constituency=&Constituency=&Constituency=&PartyFilterType=party&PoliticalParty=&PoliticalPartyGroup=&partyValueName="
    
# Create the Soup object from the html
soup = html_loader(url)
    
# Retrieve Name, Party, Region, Charge and Url
name_party_region_charge_soup = soup.find_all('div',{"class" : "person-search-result-item"})

url_soup = soup.find_all('div',{"class" : "person-search-result-item"})
url3 = 'https://senedd.wales/'

for counter1 in name_party_region_charge_soup:
    name_party_region_charge.append(counter1.text.strip())
    
for counter2 in range(len(name_party_region_charge)):
    sophie = name_party_region_charge[counter2].split('\n')
    
    if len(sophie) == 4:
        name.append(sophie[0])
        region.append(sophie[1])
        party.append(sophie[2])
        charge.append(sophie[3])
        
    if len(sophie) == 3:
        name.append(sophie[0])
        region.append(sophie[1])
        party.append(sophie[2])
        charge.append('')

print('Completed')
    
```

```python
# Retrieve Hrefs for each MP 
for counter3 in url_soup:
    urls = counter3.a.get('href')
        
    # For each Href retrieved create a new Url and convert it into a soup object using html_loader
    laida = (url3 + urls)
    soup = html_loader(laida)
    print(laida)
    
    # Retrieve Address:
    
    address_soup = soup.find_all('div',{"class" : "contact-details__address"})
    
    for counter4 in range(len(address_soup)):
        
        if ('Senedd Address:') in (address_soup[counter4].text.strip()):
            
            address.append((address_soup[counter4].text.strip()).replace("\n", " ").replace("Senedd Address: ", ""))
    
    # Retrieve Phone and Email:
    
    phone_email_soup = soup.find_all('div',{"class" : "person-details-social-item"})
    
    for counter5 in range(len(phone_email_soup)):
        
        
        if ('tel:') in (phone_email_soup[counter5].a.get('href')):
            phone.append(phone_email_soup[counter5].a.get('href').replace('tel:',''))
         
            
        if ('mailto:') in (phone_email_soup[counter5].a.get('href')):
            email.append(phone_email_soup[counter5].a.get('href').replace('mailto:',''))
            
    print(phone_email_soup)
    
   
    
```

```python
#Create a DateFrame with Pandas
df = pd.DataFrame([name, party, charge, region, phone, email, address])
df = df.transpose()
df.columns = ['Name', 'Party', 'Charge', 'Region', 'Phone', 'Email', 'Address']
df1 = df

#df = df.drop(['index'], axis=1)


#Sort by Party and export to Excel
#df1.sort_values(by='Party',ascending=True,inplace=True)
df1.to_excel("WALESGOV.xlsx",sheet_name='WALESGOV') 
df1
```
